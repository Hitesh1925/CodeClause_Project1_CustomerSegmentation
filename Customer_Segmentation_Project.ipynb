```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans
```

```python
df = pd.read_csv("mall_customers.csv")
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
X = df[['Annual Income (k$)', 'Spending Score (1-100)']]
X.head()
```

```python
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, random_state=42)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)

plt.plot(range(1, 11), wcss)
plt.xlabel('Number of Clusters (K)')
plt.ylabel('WCSS')
plt.title('Elbow Method')
plt.show()
```

```python
kmeans = KMeans(n_clusters=5, random_state=42)
df['Cluster'] = kmeans.fit_predict(X)
```

```python
# Scatter plot with clusters
plt.figure(figsize=(8,6))
sns.scatterplot(data=df, x='Annual Income (k$)', y='Spending Score (1-100)', hue='Cluster', palette='tab10', s=60)
plt.scatter(kmeans.cluster_centers_[:,0], kmeans.cluster_centers_[:,1], s=200, marker='X', edgecolor='k', label='Centroids')
plt.title('Customer Segments (K-Means)')
plt.legend()
plt.show()
```

```python
df['Cluster'].value_counts().sort_index()
```

```python
cluster_profile = df.groupby('Cluster')[['Annual Income (k$)', 'Spending Score (1-100)', 'Age']].mean().round(2)
cluster_profile['Count'] = df['Cluster'].value_counts().sort_index().values
cluster_profile
```

```python
for col in ['Annual Income (k$)', 'Spending Score (1-100)', 'Age']:
    plt.figure(figsize=(8,4))
    sns.boxplot(x='Cluster', y=col, data=df)
    plt.title(f'{col} distribution by Cluster')
    plt.show()
```

```python
df.to_csv('mall_customers_with_clusters.csv', index=False)
```
